```python
import plotly.express as px
import pandas as pd

# Load data
data = pd.read_csv('/content/owid-co2-data.csv')

```

```python
# diplay first five rows
data.head()
```

```python
print(data.columns.values)
```

```python
# Create choropleth map

#px.choropleth
fig = px.choropleth(data, locations='country', locationmode="country names",
                    color='co2',
                    hover_name='country',
                    color_continuous_scale='Viridis',
                    range_color=(0, data['co2'].max()))

# Show plot
fig.show()
```
